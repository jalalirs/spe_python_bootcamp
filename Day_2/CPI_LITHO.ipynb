To merge the datasets from different sources, it is possible to combine them by multiple ways. A brief description is provided below. 

syntax: pd.merge(left, right, how, on)

Here, left and right indicate the first and second dataframe which are required to be merged

1. how = 'outer', This allows to take the union of datasets and append them together.
2. how = 'inner', This allows to take the intersection of datasets and keep only the common values
3. how = 'left', All indexes of left dataframe are included. If there are indexes which are not present in the 'left' dataframe, they will not be included in the merged dataset
4. how = 'right', All indexes of right dataframe are included. If there are indexes which are not present in the 'right' dataframe, they will not be included in the merged dataset

<img src="imgs/joinven.png" />

The keyword 'on' decides if the dataframes are merged on a particular ID which is a unique identifier. Eg. API Number in oil and gas wells. 

```python
data_combined = pd.merge(data, data2, how='outer', on = None)
data_combined.shape
```

**Matplotlib - Python Data Visualization Library**

```python
# Default Style Changes
# https://matplotlib.org/3.1.1/users/dflt_style_changes.html
import matplotlib.style
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['image.cmap'] = 'gray'
mpl.style.use('classic')
mpl.rcParams['grid.color'] = 'k'
mpl.rcParams['grid.linestyle'] = ':'
mpl.rcParams['grid.linewidth'] = 0.5
mpl.rcParams['figure.figsize'] = [12,12]
mpl.rcParams['figure.dpi'] = 120
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.size'] = 12
mpl.rcParams['legend.fontsize'] = 'large'
mpl.rcParams['figure.titlesize'] = 'medium'
mpl.rcParams['patch.force_edgecolor'] = True
mpl.rcParams['patch.facecolor'] = 'b'
```

```python
# for inline plotting
%matplotlib inline
# for interactive plot
#%matplotlib

# creating a figure with size (8,8) in inches
plt.figure(figsize=[8,8])

# x axis: VSH
x = data['VSH (V/V)']
# y axis: depth
y = data['DEPTH (M)']
# setting x label
plt.xlabel('VSH', fontsize=15)
# setting y label
plt.ylabel('Depth (m)', fontsize=15)
# setting title
plt.title('Line Chart Example', fontsize=20)
# Line Plot
plt.plot(x, y)
plt.show()
```

```python
# Scatter Plot Example on Log-log scale

# creating a figure with size (8,8) in inches
plt.figure(figsize=[8,8])

# x axis: KLOGH
x = data['KLOGH (MD)']
# y axis: depth
y = data['SW (V/V)']

# s denotes the size of the bubbles in scatter plot
plt.scatter(x,y, s=8, label='SW') 

# setting x label
plt.xlabel('KLOGH (MD)', fontsize=15)
# setting y label
plt.ylabel('SW (V/V)', fontsize=15)
# setting title
plt.title('Scatter Plot Example', fontsize=20)
# adding ticks on both axes
plt.tick_params(axis='both', which='major', labelsize=14)
# adding legend at best figure location
plt.legend(loc='best')
# making x axis log scale
plt.xscale('log')
# making y axis log scale
plt.yscale('log')
# setting the limit for x axis
plt.xlim(10,10000)
# setting the limit for y axis
plt.ylim(0.01, 10)
```
